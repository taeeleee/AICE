# **데이터프레임**
### **1. 데이터 살펴보기**
* **칼럼(column)** : 열(세로) 레이블에 대한 정보를 보유
* **인덱스(index)** : 행(가로 row) 인덱스에 대한 정보를 보유
* **헤드 (head)** : 데이터프레임 내의 처음 n줄 데이터를 출력하는 메소드 (n을 지정해주지 않으면 5줄을 보여줌)
* **테일 (tail)** : 데이터프레임 내의 마지막 n줄 데이터를 출력하는 메소드 (n을 지정해주지 않으면 5줄을 보여줌)

### **2. 데이터프레임의 기본정보 확인하기**
* **메소드 (method)** : typically involves parentheses and can modify the object or return a new one, enabling performing operations on the object
  * `describe()`: Generates descriptive statistics of a DataFrame or Series, including count, mean, standard deviation, min, max, and quartiles
  *`info()`: Provides a concise summary of a DataFrame, including data types, non-null values, and memory usage
  *`head()`/`tail()`: Returns the first or last n rows of a DataFrame or Series, respectively

* **속성 (attribute)** : properties accessed without parentheses that provide information about the object's structure or data
  * `shape`: Returns a tuple representing the dimensions of the DataFrame (number of rows, number of columns)
  * `columns`: Returns an Index object containing the column labels
  * `index`: Returns an Index object containing the row labels
  * `dtypes`: Returns a Series with the data type of each column
  * `values`: Returns a NumPy array representation of the DataFrame
  * `T`: Returns the transpose of the DataFrame
